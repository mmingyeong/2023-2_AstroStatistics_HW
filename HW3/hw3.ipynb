```python
# Write a code to calculate the area under a give probability distribution for any given interval. 
# For Gaussian distribution with mean \mu and standard deviation of \sigma, 
# show that the integrated probably (area under the line) between \mu-\sigma and \mu+\sigma is %68.27 of the total probability.
```

```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
import math

# Define the Gaussian (normal) distribution function
def gaussian(x, mu, sigma):
    return 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-((x - mu) / sigma) ** 2 / 2)

# Parameters for the Gaussian distribution
mu = 0         # Mean
sigma = 1      # Standard deviation

# Define the integration limits
lower_limit = mu - sigma
upper_limit = mu + sigma

# Calculate the total area under the curve (from -inf to +inf)
total_area, _ = integrate.quad(gaussian, -np.inf, np.inf, args=(mu, sigma))

# Calculate the area under the curve within the interval [μ - σ, μ + σ]
area_interval, _ = integrate.quad(gaussian, lower_limit, upper_limit, args=(mu, sigma))

# Calculate the percentage of the area within the interval
percentage_within_interval = (area_interval / total_area) * 100

# Print the results
print(f"Total area under the curve: {total_area:.4f}")
print(f"Area within the interval [{lower_limit:.2f}, {upper_limit:.2f}]: {area_interval:.4f}")
print(f"Percentage within the interval: {percentage_within_interval:.2f}%")

# Plot the Gaussian distribution
x = np.linspace(-3 * sigma + mu, 3 * sigma + mu, 1000)
y = gaussian(x, mu, sigma)
plt.plot(x, y, 'r', linewidth=2)
plt.fill_between(x, 0, y, where=(x >= lower_limit) & (x <= upper_limit), alpha=0.3)
plt.title('Gaussian Distribution with Area under [μ - σ, μ + σ]')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```
